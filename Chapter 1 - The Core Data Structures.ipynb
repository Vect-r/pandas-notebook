Think of Pandas as a supercharged Excel spreadsheet that lives entirely inside Python. To use it, you only need to understand its two foundational building blocks: the Series and the DataFrame.

Every piece of data you analyze, clean, or send to a web backend will be stored in one of these two structures.

## 1. The Series (1-Dimensional Data)
A Series is essentially a single column of data. It is very similar to a standard Python list, but with one major superpower: it has an Index (labels for each row).

Let's say we are tracking the prices of a few grocery items in INR.

```python
import pandas as pd

# A simple Python list of prices
prices_list = [45, 60, 40, 120]

# Converting it into a Pandas Series
prices_series = pd.Series(prices_list)

print(prices_series)
```

Notice the numbers 0, 1, 2, 3 on the left? That is the auto-generated index. The dtype: int64 just tells us the data type is integers.

## 2. The DataFrame (2-Dimensional Data)
This is your bread and butter. A DataFrame is a full table with rows and columns. You can think of a DataFrame as a collection of multiple Series glued together.

The easiest way to build a DataFrame from scratch is by passing a Python dictionary into pd.DataFrame(). Let's create a mini-inventory dataset for a quick-commerce app.

```python
import pandas as pd

# Creating a dictionary where keys are column names, and values are the data lists
inventory_data = {
    "Item": ["Aashirvaad Atta 5kg", "Amul Taaza Milk 1L", "Britannia Bread", "Apples 1kg"],
    "Category": ["Pantry", "Dairy", "Bakery", "Produce"],
    "Price_INR": [240, 66, 45, 180],
    "In_Stock": [True, True, False, True]
}

# Converting the dictionary into a DataFrame
df = pd.DataFrame(inventory_data)

print(df)
```
