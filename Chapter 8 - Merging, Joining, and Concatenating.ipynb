# Chapter 8 - Merging, Joining, and Concatenating

This is the Pandas equivalent of SQL Joins. There are two main ways to combine DataFrames.

## 1. Concatenation (`pd.concat`) - Stacking Data

Think of this as physically stacking two identical tables on top of each other. If you have a file for Monday's deliveries and a separate file for Tuesday's deliveries, you just want to glue them together.

```python
import pandas as pd

# Monday's data
df_mon = pd.DataFrame({"order_id": [1, 2], "driver": ["Rahul", "Amit"]})
# Tuesday's data
df_tue = pd.DataFrame({"order_id": [3, 4], "driver": ["Priya", "Rahul"]})

# Stacking them vertically
df_combined = pd.concat([df_mon, df_tue], ignore_index=True)

print("--- CONCATENATED DATA ---")
df_combined
```

## 2. Merging (`pd.merge`) - Database Style Joins

This is for when you want to combine two different tables side-by-side based on a shared column (like an ID).

Let's say we have a table of Orders, but it only has a `customer_id`. We need the actual customer names for our delivery labels, which live in a separate **Customers** table.

```python
# The Customers Table (Our "Left" table)
customers = pd.DataFrame({
    "customer_id": [101, 102, 103],
    "name": ["Aarav", "Neha", "Rohan"]
})

# The Orders Table (Our "Right" table)
orders = pd.DataFrame({
    "order_id": [5001, 5002, 5003],
    "customer_id": [101, 101, 103], # Notice customer 102 didn't order anything
    "total_inr": [450, 120, 890]
})

# Merging them together!
# 'on' specifies the column they both share.
# 'how="inner"' means we only keep rows where there is a match in BOTH tables.
merged_df = pd.merge(customers, orders, on="customer_id", how="inner")

print("\n--- MERGED DATA ---")
merged_df
```

*(Pro-tip: You can change `how=` to `"left"`, `"right"`, or `"outer"` to control exactly which rows are kept when there are missing matches, exactly like SQL).*
