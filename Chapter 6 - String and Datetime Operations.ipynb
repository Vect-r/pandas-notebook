# Chapter 6 - String and Datetime Operations

Let's imagine we are intercepting raw data coming into our API from the user frontend.

```python
import pandas as pd

# Raw data coming from a frontend application
raw_api_data = {
    "search_query": ["  Amul bUtter ", "MAGGI noodles", " tata SALT  ", "eggs  "],
    "order_placed_at": ["2026-03-23 14:05:00", "2026-03-23 14:12:30", "2026-03-23 14:45:00", "2026-03-23 15:00:15"],
    "delivered_at": ["2026-03-23 14:18:00", "2026-03-23 14:30:00", "2026-03-23 14:55:00", "2026-03-23 15:22:00"]
}

df_api = pd.DataFrame(raw_api_data)
print("--- RAW API DATA ---")
df_api
```

## 1. String Operations (`.str`)
To clean up text columns in Pandas, you use the `.str` accessor. This unlocks standard Python string methods but applies them to the whole column at once.

Let's clean up those messy search queries by removing the extra spaces (stripping) and making everything lowercase so our database can read it properly.

```python
# Chaining string methods: strip spaces, then make lowercase
df_api['clean_query'] = df_api['search_query'].str.strip().str.capitalize()

print("\n--- CLEANED QUERIES ---")
df_api[['search_query', 'clean_query']]
```

## 2. Datetime Operations (`pd.to_datetime` and `.dt`)

If you look at `df_api.info()`, those timestamp columns are just plain text (strings). Pandas doesn't know they are dates yet. We have to convert them

```python
# Convert strings to actual Datetime objects
df_api['order_placed_at'] = pd.to_datetime(df_api['order_placed_at'])
df_api['delivered_at'] = pd.to_datetime(df_api['delivered_at'])
df_api
```

Once they are Datetime objects, you can do two magical things:

**A. Time Math**: You can literally subtract them to find out exactly how long a delivery took.

```python
# Calculate delivery duration
df_api['delivery_time'] = df_api['delivered_at'] - df_api['order_placed_at']
df_api
```

**B. Extracting Time Components (`.dt`):** Similar to .str, the .dt accessor lets you pull out specific parts of the date or time.

```python
# Extract just the date, or just the minute the order was placed
df_api['order_date'] = df_api['order_placed_at'].dt.date
df_api['order_minute'] = df_api['order_placed_at'].dt.minute
df_api
```
