If you ever start pulling in data from external sources or writing web scrapers to gather product information, you will quickly realize that data in the wild is a total mess. Systems crash, fields get left blank, and web pages return weird formats.

Pandas provides incredible tools to clean up this garbage so it doesn't break your backend logic.

## Step 1: The Messy Dataset
Let's create a scraped inventory list that has some serious problems: missing values (NaN - Not a Number) and accidental duplicate entries. Run this code to see the mess:

```python
import pandas as pd

# 'None' will automatically be converted to 'NaN' (missing data) by Pandas
scraped_data = {
    "Item_ID": [101, 102, 103, 101, 104, 105],
    "Product_Name": ["Amul Milk", "Britannia Bread", None, "Amul Milk", "Eggs", "Butter"],
    "Price_INR": [66, 45, 80, 66, None, 58],
    "Stock_Count": [50, 20, 0, 50, 30, None]
}

df_messy = pd.DataFrame(scraped_data)
print("--- ORIGINAL MESSY DATA ---")
print(df_messy)
```

## Step 2: Finding the Problems
Before you fix the data, you need to know where the holes are.

`df.isna().sum()`: This is a lifesaver. It goes through every column and counts exactly how many missing values there are.

`df.duplicated().sum()`: Tells you how many exact duplicate rows exist.

```python
print("\n--- MISSING VALUES COUNT ---")
print(df_messy.isna().sum())
```

## Step 3: Fixing Missing Data
You basically have two choices when dealing with a missing value: Drop it or Fill it.

### 1. Dropping Data:
If a row is missing crucial information (like the price), you might just want to delete the whole row.

```python
# Drops ANY row that has at least one missing value
df_cleaner = df_messy.dropna() 

# Drops a row ONLY if a specific column is missing data
df_cleaner = df_messy.dropna(subset=['Price_INR'])
df_cleaner
```

### 2. Filling Data
If a product is missing its stock count, deleting the product might be a bad idea. Instead, it is safer to assume the stock is 0.

```python
# Fills missing values in a specific column with a 0
df_cleaner['Stock_Count'] = df_messy['Stock_Count'].fillna(0)

# Fills missing text data
df_cleaner['Product_Name'] = df_messy['Product_Name'].fillna("Unknown Product")
df_cleaner
```

### Step 4: Removing Duplicates
Web scrapers and user forms submit duplicate data all the time. Pandas can wipe them out in one line.

```python
# Keeps the first instance and drops the identical copies
df_unique = df_cleaner.drop_duplicates()
df_unique
```
