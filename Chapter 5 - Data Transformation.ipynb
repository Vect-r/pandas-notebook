Data Transformation is all about modifying your existing data or creating new, useful data out of it.

Let's use a fresh, clean dataset of an e-commerce inventory for this:

```python
import pandas as pd

inventory_data = {
    "product_sku": ["SKU-101", "SKU-102", "SKU-103", "SKU-104"],
    "name": ["Aashirvaad Atta 5kg", "Amul Butter 100g", "Eggs 6pcs", "Maggi 2-Min Noodles"],
    "price": [240, 58, 42, 14],
    "stock_qty": [15, 2, 0, 50]
}

df_inv = pd.DataFrame(inventory_data)
```

## 1. Creating Calculated Columns

You can easily create new columns by doing math on existing columns. This is highly optimized in Pandas (Vectorization), meaning it calculates the whole column at once, without a for loop.

Let's calculate the total potential revenue of the inventory we currently hold:

```python
# Creates a new column by multiplying price and stock_qty
df_inv['total_value_inr'] = df_inv['price'] * df_inv['stock_qty']

df_inv[['name', 'total_value_inr']]
```

## 2. The Magic of `.apply`()

Sometimes, simple math isn't enough. You might need to run custom logic on every row. The `.apply()` method lets you pass a Python function to your data.

Let's say we want a column that flags whether an item needs to be restocked.

```python
# First, write a standard Python function
def check_stock(qty):
    if qty == 0:
        return "OUT OF STOCK"
    elif qty < 5:
        return "LOW STOCK"
    else:
        return "IN STOCK"

# Now, apply this function to the 'stock_qty' column
df_inv['stock_status'] = df_inv['stock_qty'].apply(check_stock)

df_inv[['name', 'stock_qty', 'stock_status']]
```

## 3. Renaming Columns

If you are passing this data to a REST API, you might want your column names to match your Django models or frontend expectations perfectly.

```python
# Pass a dictionary where { "old_name" : "new_name" }
df_inv = df_inv.rename(columns={
    "product_sku": "sku",
    "name": "product_name"
})
df_inv
```
