# Chapter 3: Selection and Filtering

```python
import pandas as pd

df = pd.read_csv("datasets\\quick_commerce_orders.csv")
```

## 1. Selecting Columns (Slicing the Data)
Sometimes you don't need the whole table. You just want to isolate one or two specific columns.

Selecting a Single Column:
When you select a single column, Pandas returns it as a Series (1D data). You just pass the column name inside square brackets.

```python
# Let's look at just the items ordered
items = df['Item_Name']
print(items.head())
```

### Selecting Multiple Columns:
To select multiple columns, you pass a list of column names. Notice the double square brackets [[ ]]! The inner brackets are the Python list, and the outer brackets are the Pandas selector. This returns a new, smaller DataFrame.

```python
# Creating a streamlined view for the delivery drivers
driver_view = df[['Order_ID', 'Delivery_Time_Min', 'Delivered_On_Time']]
print(driver_view.head())
```

## 2. Filtering Rows (Conditional Logic)
This is where Pandas shines. You can filter rows based on conditions using a syntax that feels very similar to standard Python, but applied to an entire column at once.

Let's find all the orders that were not delivered on time.

```python
# 1. Create the condition (This creates a Series of True/False values)
late_condition = df['Delivered_On_Time'] == False

# 2. Pass that condition into the DataFrame brackets to filter the rows
late_orders = df[late_condition]

print("--- LATE ORDERS ---")
print(late_orders[['Order_ID', 'Item_Name', 'Delivery_Time_Min']])

df[df['Delivered_On_Time']==False][['Order_ID','Item_Name','Delivery_Time_Min']]
```

> Pro-tip: Most developers write this in a single line like this: late_orders = df[df['Delivered_On_Time'] == False]

## 3. Multiple Conditions (AND / OR)
You can combine conditions to get highly specific data.

Use & for AND (both conditions must be true).

Use | for OR (either condition can be true).
(Note: You must wrap each individual condition in parentheses () when combining them).

Let's find high-value orders (Total Price > ₹100) specifically in the "Dairy" category.

```python
high_value_dairy = df[(df['Total_Price_INR'] > 100) & (df['Category'] == 'Dairy')]

print("--- HIGH VALUE DAIRY ORDERS ---")
high_value_dairy
```
