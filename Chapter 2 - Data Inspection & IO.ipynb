## 2. Inspecting Your Data
When you load a massive dataset, you can't just print(df) because it will flood your terminal. Instead, you need tools to get a quick summary.

Imagine we loaded a dataset of 10,000 grocery deliveries. Here are the four most important commands you will use every single day:

`df.head()` : Shows the first 5 rows of your data. Great for a quick visual check. (You can also pass a number, like df.head(10)).

`df.tail()` : Shows the last 5 rows.

`df.info()` : Gives you a technical summary of your DataFrame. It tells you the number of rows, the names of the columns, if there is any missing data (Null values), and what data type each column holds (e.g., integers, strings/objects, booleans).

`df.describe()` : This is the magic command for data analytics. It instantly calculates statistics for all numerical columns (count, mean, standard deviation, minimum, maximum, and percentiles).
